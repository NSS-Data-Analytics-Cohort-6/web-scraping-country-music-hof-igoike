```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as BS
import numpy as np
from requests import get
import matplotlib.pyplot as plt
import re
%matplotlib inline
```

```python
#full code to extract all the date (polished version)

page = 1
scraped_data = []
while page!=18:
      url= f'https://countrymusichalloffame.org/hall-of-fame/members/page/{page}'
      response = requests.get(url)
      data = BS(response.text, 'html.parser')
      cards_data = data.find_all('div', attrs={'class', 'vertical-card_content'})  
      for card in cards_data:
          card_details = {}
          links_data = card.find('a')
          link_temp = str(links_data.attrs['href'])    
          names = card.find('h3')
          year_inducted = str(card.find('p'))
          year_new = re.findall(r'\d',year_inducted)
          year = ''.join(year_new)
          url_new= link_temp
          response_new = requests.get(url_new)                   
          sub_data = BS(response_new.text, 'html.parser')
          d = str(sub_data.find_all('p'))         
          birth_day = re.findall(r'(?<=Birth:)\s\w+\s\d+,\s\d+',d)
          birth_day_new = ",".join(birth_day)  
          death_day = re.findall(r'(?<=Death:)\s\w+\s\d+,\s\d+',d)
          death_day_new = ",".join(death_day)  
          birth_place = str(re.findall(r'(?<=Birthplace:)\s([^,]+),(\s\w+)',d))         
          birth_place_new = str(re.findall(r'\w+',birth_place))
          birth_place_new = birth_place_new.replace('['," ")
          birth_place_new = birth_place_new.replace(']'," ")
          birth_place_new = birth_place_new.replace("'","")  
          card_details['name'] = names.text
          card_details['year_inducted'] = year
          card_details['birth day'] = birth_day_new  
          card_details['died'] = death_day_new
          card_details['birth_place'] = birth_place_new 
          scraped_data.append(card_details)           
          print(page,names,year,birth_day_new,death_day_new,birth_place_new)   
      page = page + 1      
      df_hall_of_fame = pd.DataFrame.from_dict(scraped_data)
```
